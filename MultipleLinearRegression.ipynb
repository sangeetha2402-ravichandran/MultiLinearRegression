STEP 4 CREATE TRAINING AND TESTING DATA

```python
y = stock['S&P 500 Price']
```

```python
X = stock[['Interest Rates', 'Employment']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
y_test.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

STEP 5 TRAIN MODEL

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression(fit_intercept = True) # False then  b is 0
regression.fit(X_train, y_train)
```

```python
print('Linear regression coefficient m- slope', regression.coef_)
print('Linear regression coefficient b - intercept', regression.intercept_)
```

STEP 6 EVALUATE THE MODEL

```python
y_predict = regression.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
plt.scatter(y_predict, y_test, color = 'gray')
plt.ylabel("Model prediction")
plt.xlabel("True Values")
plt.title("MultiLinearRegression")
```

```python
k=X_test.shape
n=len(X_test)
```

```python
k
```

```python
n
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
RMSE = float(format(np.sqrt(mean_squared_error(y_test,y_predict)),'.3f')) # root mean square
MSE= mean_squared_error(y_test,y_predict) # mean square error
MAE= mean_absolute_error(y_test,y_predict) # mean aboslute error
r2 = r2_score(y_test,y_predict) # r square
NPSS = np.mean(np.abs((y_test - y_predict) / y_test) ) * 100 # mean absolute percentage of error
print('RMSE=', RMSE, '\nMSE=' , MSE , '\nMAE=' , MAE , '\nr2=',  r2 ,'\nNPS=' , NPS)
             
```
